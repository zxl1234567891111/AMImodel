```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set Chinese font display (if needed to display Chinese labels)
plt.rcParams['font.sans-serif'] = ['SimHei']  # For displaying Chinese labels
plt.rcParams['axes.unicode_minus'] = False  # For displaying negative signs

# Read Excel file
file_path = 'ALL.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Check data
print("First 5 rows of data:")
print(data.head())
print("\nData shape:", data.shape)

# Calculate correlation matrix
corr_matrix = data.corr()

# Set figure size
plt.figure(figsize=(20, 16))

# Draw heatmap
heatmap = sns.heatmap(corr_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            center=0,
            linewidths=0.5,
            annot_kws={"size": 15})  # Adjust font size of values in heatmap

# Adjust colorbar font size
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=20)  # Adjust colorbar tick font size

# If you need to adjust colorbar title font size (if present)
# cbar.set_label('Correlation Coefficient', fontsize=16)

# Adjust labels and title
# plt.title('Myocardial Infarction Feature Correlation Heatmap', fontsize=20, pad=20)
plt.xticks(rotation=45, ha='right', fontsize=20)
plt.yticks(fontsize=20)

# Automatically adjust layout to prevent label cutoff
plt.tight_layout()

# Save image
plt.savefig('myocardial_infarction_feature_correlation_heatmap.png', dpi=300, bbox_inches='tight')

# Display image
plt.show()
```
